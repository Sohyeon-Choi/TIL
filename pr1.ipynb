## Grouping

```python
In [87]: df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)

In [88]: df
```

```python
In [89]: df.groupby("A")[["C", "D"]].sum()
```

```python
In [90]: df.groupby(["A", "B"]).sum()
```
